```python
import os
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

X = [[1], [2], [9], [12], [20]]

#Delete the existing png file
if os.path.exists("AHC_X_single.png"):
    os.remove("AHC_X_single.png")

#AHC clustering using Single Linkage on X dataset
Z = linkage(X, 'single')

# Plot the dendrogram for the single linkage
dendrogram(Z)

#Save the plot as a png file
plt.savefig('AHC_X_single.png')
plt.title('AHC on X dataset using Single Linkage')
plt.show()
"""
   Explain the dendrogram plot for single linkage
    The dendrogram plot for single linkage shows the distance between the clusters.
    The distance between the clusters is shown on the y-axis and the data points are shown on the x-axis.
    The dendrogram plot shows that the distance between the clusters is increasing as we move up the y-axis.
""" 
```

```python
#Delete the existing png file if it exists
if os.path.exists("AHC_X_complete.png"):
    os.remove("AHC_X_complete.png")

#AHC clustering using Complete Linkage on X dataset
Z = linkage(X, 'complete')

# Plot the dendrogram for the Complete linkage
dendrogram(Z)

#Save the plot as a png file
plt.savefig('AHC_X_complete.png')
plt.title('Dendrogram for Complete Linkage on X dataset')
plt.show()
""" 
    Explain the dendrogram plot for complete linkage
    The dendrogram plot for complete linkage shows the distance between the clusters.
    The distance between the clusters is shown on the y-axis and the data points are shown on the x-axis.
    The dendrogram plot shows that the distance between the clusters is increasing as we move up the y-axis.
    The dendrogram plot for complete linkage shows that the distance between the clusters is more than the distance between the clusters in single linkage.
""" 
```

```python
# Dataset 1: Iris
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
import os

iris = datasets.load_iris()
X_ir = iris.data
y_ir = iris.target

#Delete the existing png file if it exists
if os.path.exists("AHC_iris_single.png"):
    os.remove("AHC_iris_single.png")

#AHC clustering using Single Linkage on Iris dataset
A = linkage(X_ir, 'single')

# Plot the dendrogram for the single linkage
dendrogram(A)

#Save the plot as a png file
plt.savefig('AHC_iris_single.png')
plt.title("AHC clustering using Single Linkage on Iris dataset")
plt.show()
```

```python
#Delete the existing png file if it exists
if os.path.exists("AHC_iris_complete.png"):
    os.remove("AHC_iris_complete.png")

#AHC clustering using Complete Linkage on Iris dataset
B = linkage(X_ir, 'complete')

# Plot the dendrogram for the Complete linkage
dendrogram(B)

#Save the plot as a png file
plt.savefig('AHC_iris_complete.png')
plt.title("AHC clustering using Complete Linkage on Iris dataset")
plt.show()
```
